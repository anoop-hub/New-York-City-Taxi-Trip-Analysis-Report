```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Data Manipulation and Understanding
#- Utilize the Pandas library in Python to load and conduct an initial exploration of the NYC Taxi Trip Duration dataset: 
#- Summarize your findings, focusing on the dataset's structure, presence of missing values, and the data types of various columns.
```

```python
# Load the dataset
train_data_path = (r"C:\Users\Lenovo\Downloads\nyc-taxi-trip-duration/train.csv")
test_data_path = (r"C:\Users\Lenovo\Downloads\nyc-taxi-trip-duration/test.csv")
sample_submission_path = (r"C:\Users\Lenovo\Downloads\nyc-taxi-trip-duration/sample_submission.csv")

# Reading the datasets
train_df = pd.read_csv(train_data_path)
test_df = pd.read_csv(test_data_path)
sample_submission_df = pd.read_csv(sample_submission_path)

# Display the first few rows of the datasets
train_df.head(), test_df.head(), sample_submission_df.head()

# Exploring the dataset's structure, missing values, and data types
print("\nDataset Info:")
print(train_df.info())

# Checking for missing values
print("\nMissing Values:")
print(train_df.isnull().sum())
```

```python
#2.Data Visualization and Analysis
#Create three distinct types of graphs using Matplotlib to highlight key observations from the dataset.
#For each graph:
#- Justify your choice of graph type for the data being represented.
#- Discuss the insights or trends that the graph uncovers about the dataset.
#- Ensure the graph is properly annotated with a clear title, axis labels, and a legend (if necessary).
```

```python
# Define the graphs to be created based on the initial exploration

# Graph 1: Histogram of Trip Durations
# Justification: To understand the distribution of trip durations
plt.figure(figsize=(10, 6))
sns.histplot(train_df['trip_duration'], bins=50, kde=True)
plt.title('Distribution of Trip Durations')
plt.xlabel('Trip Duration (seconds)')
plt.ylabel('Frequency')
plt.xlim(0, 10000)  # Limiting to 10000 seconds for better visualization
plt.grid(True)
plt.show()

# Graph 2: Scatter plot of Pickup Latitude and Longitude
# Justification: To visualize the geographical distribution of taxi pickups
plt.figure(figsize=(10, 6))
plt.scatter(train_df['pickup_longitude'], train_df['pickup_latitude'], alpha=0.1)
plt.title('Geographical Distribution of Taxi Pickups')
plt.xlabel('Pickup Longitude')
plt.ylabel('Pickup Latitude')
plt.grid(True)
plt.show()

# Graph 3: Boxplot of Passenger Count vs Trip Duration
# Justification: To observe how trip duration varies with different passenger counts
plt.figure(figsize=(10, 6))
sns.boxplot(x='passenger_count', y='trip_duration', data=train_df)
plt.title('Passenger Count vs Trip Duration')
plt.xlabel('Passenger Count')
plt.ylabel('Trip Duration (seconds)')
plt.ylim(0, 5000)  # Limiting trip duration for better visualization
plt.grid(True)
plt.show()
```
